```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
def n_squared(lam: np.ndarray, lambdas: np.ndarray, gs: np.ndarray):
    '''sellmayer eq from culi et at 2003, eq (13)

    Parameters
    ----------
    lam : _type_
        wavelength
    lambdas : np.ndarray
        vector of [lambda_1, lambda_2, lambda_3 ]
    gs : np.ndarray
        vector of [g_0, g_1, g_2, g_3]
    '''
    zero = gs[0]
    others = np.array([gs[i]/(lambdas[i]**(-2) - lam**(-2)) for i in range(3)])
    return zero + np.sum(others, axis=0)
```

```python
lam = np.linspace(0.95, 15, 100)  # wavelength in um

# values from table II
lambdas = [0.4431307,
           0.8746453,
           36.9166]

gs = [5.372514,
      27.83972,
      0.031764,
      0.001]
```

```python
n = np.sqrt(n_squared(lam, lambdas, gs))
```

```python
plt.figure(dpi=250, facecolor='w')
plt.plot(lam, n)
plt.xlabel('$\lambda$ ($\mu$m)')
plt.ylabel('n')
plt.grid()
```

> Am i seeing typical behavior?

Yes I am. On the left the refractive index rises as it is converging to the resonance freq.
> Normal or anomalouse disp?

Shorter wavelengths (higher freq.) have higher refr. index and therefore the travel time. -> the dispersion is **normal**
